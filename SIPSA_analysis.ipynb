#**Inflación Análisis**

Precios de los Alimentos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("/content/Historico SIPSA.xlsx", sheet_name= "Enero")
Enero = df.rename(columns={'Precio promedio por kilogramo*':'Precio'}).drop(['Enero de 2023'], axis=1)
Enero
```

```python
Enero.info()
```

```python
Enero['Grupo'].unique()
```

```python
Categorias =df.groupby(['Grupo', 'Producto'])['Precio promedio por kilogramo*'].count()
Categorias
```

```python
Carnes = Enero[Enero['Grupo'] == 'CARNES']
Verduras = Enero[Enero['Grupo'] == 'VERDURAS Y HORTALIZAS']
Frutas = Enero[Enero['Grupo'] == 'FRUTAS']
GRANOS_CEREALES = Enero[Enero['Grupo'] == 'GRANOS Y CEREALES']
Lacteos_huevos= Enero[Enero['Grupo'] == 'LACTEOS Y HUEVOS']
```

```python
promedio_carnes = Carnes['Precio'].mean()
promedio_verduras = Verduras['Precio'].mean()
promedio_frutas = Frutas['Precio'].mean()
promedio_GRANOS_CEREALES= GRANOS_CEREALES['Precio'].mean()
promedio_Lacteos_huevos= Lacteos_huevos['Precio'].mean()
```

```python
print(f'Promedio de precios para carnes:{promedio_carnes}')
print(f'Promedio de precios para verduras:{promedio_verduras}')
print(f'Promedio de precios para frutas:{promedio_frutas}')
print(f'Promedio de precios para Granos y cereales:{promedio_GRANOS_CEREALES}')
print(f'Promedio de precios para Lacteos y huevos:{promedio_Lacteos_huevos}')
```

```python
#Exportar como CSV
promedios_df = pd.DataFrame({'Subgrupo': ['Carnes','Verduras','Frutas','GRANOS_CEREALES','Lacteos_huevos'],
    'Precio_Promedio_Enero': [promedio_carnes,promedio_verduras, promedio_frutas, promedio_GRANOS_CEREALES,promedio_Lacteos_huevos]})
```

```python
promedios_df.to_csv('Precio_Promedio_Enero.csv', index=False)
```

```python
grupos = ['Carnes','Verduras','Frutas']
promedios = [promedio_carnes,promedio_verduras, promedio_frutas]
```

```python
plt.bar(grupos, promedios, color=['green', 'orange', 'blue'])
plt.title('Promedio de Precios por Subgrupo')
plt.xlabel('Subgrupo')
plt.ylabel('Precio Promedio')
plt.show()
```
